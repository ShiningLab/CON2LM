# Main Results

```python
# Table 1: Accuracy (%) on contradiction detection
# Data from the paper - TEMP context, Direct surprisal, Mean aggregation
accuracy_data = {
    'models': ['Length', 'Llama', 'Llama-I', 'Gemma', 'Qwen'],
    'Val': [52.5, 63.9, 63.7, 66.2, 66.5],
    'SNLI': [50.1, 62.4, 61.5, 63.6, 64.3],
    'bAbI': [45.0, 59.2, 61.7, 54.2, 72.5],
    'Wiki-C': [50.0, 97.0, 93.0, 97.0, 99.0],
    'Wiki-L': [50.0, 91.0, 90.0, 85.0, 99.0],
    'Wiki-S': [50.0, 97.0, 99.0, 83.0, 100.0]
}
# Table 2: ROC-AUC (%) measuring correlation between surprisal and contradiction
roc_auc_data = {
    'models': ['Length', 'Llama', 'Llama-I', 'Gemma', 'Qwen'],
    'Val': [48.3, 69.0, 68.5, 71.1, 71.1],
    'SNLI': [47.5, 66.7, 66.8, 69.1, 69.6],
    'bAbI': [52.7, 62.5, 64.7, 56.1, 86.9],
    'Wiki-C': [53.3, 100.0, 99.6, 100.0, 98.1],
    'Wiki-L': [49.4, 100.0, 100.0, 100.0, 100.0],
    'Wiki-S': [51.7, 100.0, 100.0, 100.0, 100.0]
}
```

```python
# Create grouped bar charts for Accuracy and ROC-AUC
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6), dpi=100)

# Define datasets (excluding Val which is just for tuning)
datasets = ['SNLI', 'bAbI', 'Wiki-C', 'Wiki-L', 'Wiki-S']
n_datasets = len(datasets)
n_models = 5  # Length, Llama, Llama-I, Gemma, Qwen

# Bar width and positions
bar_width = 0.15
x_pos = np.arange(n_datasets)

# ========== ACCURACY CHART (Table 1) ==========
for i, model in enumerate(accuracy_data['models']):
    values = [accuracy_data[dataset][i] for dataset in datasets]
    offset = (i - 2) * bar_width  # Center the bars around 0
    
    ax1.bar(
        x_pos + offset,
        values,
        width=bar_width,
        color=colors[i * 2],
        edgecolor='black',
        linewidth=2,
        label=model
    )

ax1.set_xticks(x_pos)
ax1.set_xticklabels(datasets, fontsize=22)
ax1.set_ylabel('Accuracy (%)', fontsize=26)
ax1.set_ylim([40, 105])
ax1.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax1.set_axisbelow(True)

# After creating the accuracy bars (ax1)
ax1.axhline(y=50.1, color='red', linestyle='--', linewidth=2, alpha=0.7)

# ========== ROC-AUC CHART (Table 2) ==========
for i, model in enumerate(roc_auc_data['models']):
    values = [roc_auc_data[dataset][i] for dataset in datasets]
    offset = (i - 2) * bar_width  # Center the bars around 0
    
    ax2.bar(
        x_pos + offset,
        values,
        width=bar_width,
        color=colors[i * 2],
        edgecolor='black',
        linewidth=2,
        # label=model
    )

ax2.set_xticks(x_pos)
ax2.set_xticklabels(datasets, fontsize=22)
ax2.set_ylabel('ROC-AUC (%)', fontsize=26)
ax2.set_ylim([40, 105])
ax2.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax2.set_axisbelow(True)

# Shared legend at the top
fig.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, 1.09),
    ncol=5,
    fancybox=True,
    shadow=True,
    fontsize=24
)

# After creating the ROC-AUC bars (ax2)
ax2.axhline(y=53.3, color='red', linestyle='--', linewidth=2, alpha=0.7)

plt.tight_layout()
plt.savefig('res/figures/results.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```
